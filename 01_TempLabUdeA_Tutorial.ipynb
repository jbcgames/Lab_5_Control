# Visión general de TCLab

El paquete `tclab` ofrece un conjunto de herramientas en Python para interactuar con el [Laboratorio de Control de Temperatura de BYU](http://apmonitor.com/pdc/index.php/Main/ArduinoTemperatureControl). Este laboratorio consta de dos calentadores y dos sensores de temperatura montados sobre una placa de Arduino. En conjunto, el paquete `tclab` y el Laboratorio de Control de Temperatura proporcionan una plataforma experimental de bajo costo para implementar algoritmos de control de procesos.

## Arquitectura de TCLab

El paquete `tclab` ha sido concebido como una **herramienta didáctica** para el estudio de sistemas de control de procesos. Proporciona una interfaz de **alto nivel** para:

- **Sensores de temperatura**  
- **Elementos calefactores**  
- **Reloj pseudo–tiempo real**  
- **Registro y visualización de datos**

En concreto, `tclab` incluye las siguientes clases y utilidades en Python:

- `TCLab()`  
  Permite el acceso directo al hardware del Laboratorio de Control de Temperatura real, gestionando la comunicación serie con la placa Arduino y los módulos de calefacción y medición.

- `TCLabModel()`  
  Ofrece un modelo de simulación del laboratorio, ideal para pruebas preliminares o desarrollo de algoritmos sin disponer del dispositivo físico.

- `clock`  
  Un mecanismo de sincronización que emula un reloj de tiempo real, facilitando la implementación de experimentos con temporización precisa.

- `Historian`  
  Componente para almacenamiento estructurado de datos de medición y control a lo largo del experimento.

- `Plotter`  
  Herramienta de trazado en tiempo real de variables de proceso (temperaturas, potencias, etc.) directamente en Jupyter Notebooks o scripts interactivos.

TCLabOverview - copia.png

Gracias a este conjunto de recursos, los estudiantes pueden desarrollar notebooks y programas en Python que abarquen desde **identificación de modelos**, **tuning de controladores PID**, hasta **estrategias avanzadas de control predictivo**.

Adicionalmente, el proyecto incluye:

- **tclab.py**  
  Librería principal en Python que expone las clases `TCLab`, `TCLabModel`, `clock`, `Historian` y `Plotter`.

- **TempLAB_UdeA_Arduino.ino**  
  Firmware para la operación segura y determinista de la placa Arduino y el shield de laboratorio. Disponible en  
  <https://drive.google.com/file/d/1atRzAmglky8Ly4mfhQD6dycIr6AMyDOg/view?usp=sharing>.

- **Arduino**  
  Plataforma de hardware compatible (Arduino Uno, Leonardo y clones) sobre la cual corre el firmware y se montan sensores y actuadores del laboratorio.


<h2>Comenzando</h2>

<h3>Instalación</h3>
<pre><code>pip install tclab</code></pre>
<p>Para actualizar una instalación existente, use:</p>
<pre><code>pip install tclab --upgrade</code></pre>


<h3>Configuración de hardware</h3>
<ol>
  <li>Conecte un dispositivo Arduino compatible (UNO, Leonardo, NHduino) con el laboratorio acoplado a su equipo mediante USB. Conecte el adaptador de corriente continua a la red eléctrica.</li>
  <li><strong>(Opcional)</strong> Instale los controladores de Arduino.<br>
    <em>Nota:</em> En Windows 10 normalmente no se requieren controladores adicionales.  
    En macOS, si usa clones basados en CHIP CH340, CH34G o CH34X, puede encontrar el driver  
    <a href="https://github.com/MPParsley/ch340g-ch34g-ch34x-mac-os-x-driver">aquí</a> o  
    <a href="https://github.com/adrianmihalko/ch340g-ch34g-ch34x-mac-os-x-driver">aquí</a>.
  </li>
  <li><strong>(Opcional)</strong> Instale el firmware de Arduino;<br>
    <code>TCLab</code> requiere cargar una sola vez un firmware personalizado en la placa Arduino.  
    Si aún no está instalado, el firmware y las instrucciones están disponibles en  
    <a href="https://drive.google.com/file/d/1atRzAmglky8Ly4mfhQD6dycIr6AMyDOg/view?usp=sharing">TempLabUdeA_Arduino</a>.
  </li>
</ol>

<h3>Verificación de funcionamiento</h3>
<p>Ejecute el siguiente fragmento en Python:</p>
<pre><code>import tclab
with tclab.TCLab() as lab:
    print(lab.T1)</code></pre>
<p>Si todo ha funcionado correctamente, debería verse una salida similar a:</p>
<pre><code>Connecting to TCLab
TCLab Firmware Version 1.2.1 on NHduino connected to port XXXX
21.54
TCLab disconnected successfully.</code></pre>
<p>El valor devuelto corresponde a la temperatura del sensor <strong>T1</strong> en grados Celsius.</p>
